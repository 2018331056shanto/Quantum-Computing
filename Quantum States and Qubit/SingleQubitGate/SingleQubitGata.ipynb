## The Pauli Gates




### The X-Gate

#### The X-gate is represented by the Pauli-X matrix:
$X=\begin{bmatrix}
0 & 1\\
1 & 0
\end{bmatrix}=|\left. 0 \right\rangle \left\langle 1| \right. + |\left. 1 \right\rangle \left\langle 0| \right.$

### Outer Product

$\begin{bmatrix}
a \\
b 
\end{bmatrix}$
$\begin{bmatrix}
c \\
d 
\end{bmatrix}=
\begin{bmatrix}
ac & ad \\
bc & bd
\end{bmatrix}$

#### To see the effect a gate has on a qubit, we simply multiply the qubit’s statevector by the gate. We can see that the X-gate switches the amplitudes of the states $\left. |0\right\rangle$ and  $\left. |1\right\rangle$:

$X \left.|0\right\rangle=\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
\begin{bmatrix}
1 \\
0 
\end{bmatrix}=\begin{bmatrix}
0 \\
1
\end{bmatrix}=\left. |1\right\rangle$

#### Another way of representing using **Bra-Ket** notation

$X=\left.|0\right\rangle \left\langle 1|\right. + \left.|1\right\rangle \left\langle 0|\right.$

$X\left. |1\right\rangle=(\left.|0\right\rangle \left\langle 1|\right. + \left.|1\right\rangle \left\langle 0|\right)\left. |1\right\rangle$

$=\left. |0\right\rangle \left\langle 1|1 \right\rangle + \left. |1\right\rangle \left\langle 0|1 \right\rangle $

$=\left. |0 \right\rangle *1 + \left. |1\right\rangle *0$

$=\left. |0\right\rangle$

$\left\langle \psi|\psi \right\rangle =1 \left\langle \psi|orthonormal basis \right\rangle=0$

### Eigenvectors of Matrices

#### We have seen that multiplying a vector by a matrix results in a vector:
$M\left.|v\right\rangle= \left. |v\prime\right\rangle \leftarrow New Vector$

#### If we chose the right vectors and matrices, we can find a case in which this matrix multiplication is the same as doing a multiplication by a scalar:

$M\left.|v\right\rangle= \lambda\left. |v\right\rangle $

#### Here V is eigenvector of M and $\lambda$ is eigenvalue

#### The eigenvectors of the Z-matrix are the states $\left.|0\right\rangle$ and $\left.|1\right\rangle$

$Z\left.|0\right\rangle=\left.|0\right\rangle$

$Z\left.|1\right\rangle=-\left.|1\right\rangle$

#### Since we use vectors to describe the state of our qubits, we often call these vectors eigenstates in this context

#### The states $\left.|0\right\rangle$ and $\left.|1\right\rangle$  are the two eigenstates of the Z-gate. In fact, the computational basis (the basis formed by the states  $\left.|0\right\rangle$  and   $\left.|1\right\rangle$ ) is often called the Z-basis.

#### A popular basis is the X-basis, formed by the eigenstates of the X-gate

$\left.|+\right\rangle=\frac{1}{\sqrt{2}}(\left.|0\right\rangle + \left. |1\right\rangle)=\begin{bmatrix}
1 \\
1
\end{bmatrix}$

$\left.|-\right\rangle=\frac{1}{\sqrt{2}}(\left.|0\right\rangle - \left. |1\right\rangle)=\begin{bmatrix}
1 \\
-1
\end{bmatrix}$

## The Hadamard Gate


#### Using only the Pauli-gates it is impossible to move our initialized qubit to any state other than $\left. |0\right\rangle$ or $\left.|1\right\rangle$ , i.e. we cannot achieve superposition

#### **The Hadamard Gate** allows us to move away from the poles of the Bloch sphere and create a superposition of $\left.|0\right\rangle$ and $\left.|1\right\rangle$. It has the matrix:

$H=\frac{1}{\sqrt{2}}\begin{bmatrix}
1 & 1\\
1 & -1
\end{bmatrix}$

#### We can see that this performs the transformations below:

$H\left.|0\right\rangle =\left.|+\right\rangle$

$H\left.|1\right\rangle =\left.|-\right\rangle$

## Digression: Measuring in Different Bases

#### Z-axis is not intrinsically special, and that there are infinitely many other bases. Similarly with measurement, we don’t always have to measure in the computational basis (the Z-basis), we can measure our qubits in any basis.

#### Measuring in the X-basis. We can calculate the probability of measuring a qubit state either $\left. |+\right\rangle$ or $\left. |-\right\rangle$  with any orthonormal bases. Suppose we are measuring in **q** bases :

$p(\left.|+\right\rangle)=|\left\langle+|q\right\rangle|^2$

$p(\left.|-\right\rangle)=|\left\langle-|q\right\rangle|^2$

#### We can choose to measure a qubit in some basis that means we chooose one of these pairs of orthogonal states of basis states and we're going to measure the state of the qubit relative to that particular pair of state.




```python
qc=QuantumCircuit(1)
initial_state=[1/sqrt(2),-1/sqrt(2)]
qc.initialize(initial_state,0)

qc.save_statevector()
qc.draw()
```

```python
sim=Aer.get_backend('aer_simulator')
results=sim.run(qc).result()
output=results.get_statevector()
plot_bloch_multivector(output)
```

#### Here we are measuring the state of a qubit in X-basis. Which means we want to means the state of a qubit relative to the X-basis . When we measure a state collapse to that state  in which basis we measured. Here if the initial sate of the qubit is $\left.|-\right\rangle$ then if we measure in X-basis we are certain that after measuring we will get either $1$ if the initial state were $\left.|+\right\rangle$ we would get $0$  

```python
# Creating X-measurement function
def x_measurement(qc,qbit,cbit):
    qc.h(qbit)
    qc.measure(qbit,cbit)
    return qc
```

#### Initial State $\left.|-\right\rangle$

```python
qc=QuantumCircuit(1,1)
initial_state=[1/sqrt(2),-1/sqrt(2)]
qc.initialize(initial_state,0)
x_measurement(qc,0,0)
qc.draw()
```

```python
results=sim.run(qc).result()

counts=results.get_counts()
plot_histogram(counts)
```

#### Initial State $\left.|+\right\rangle$

```python
qc=QuantumCircuit(1,1)
initial_state=[1/sqrt(2),1/sqrt(2)]
qc.initialize(initial_state,0)
x_measurement(qc,0,0)
qc.draw()
```

```python
results=sim.run(qc).result()

counts=results.get_counts()
plot_histogram(counts)
```

#### Now if the initial state is either $\left.|0\right\rangle$ or $\left.|1\right\rangle$ then according to the X-basis they are in superposition state. So we will have $50/50$ chance of getting either $0$ or $1$

```python
qc=QuantumCircuit(1,1)
initial_state=[1,0]
qc.initialize(initial_state,0)
x_measurement(qc,0,0)
qc.draw()
```

```python
results=sim.run(qc).result()

counts=results.get_counts()
plot_histogram(counts)
```

## The P-gate

####  The P-gate performs a rotation of $(\phi)$ around the Z-axis direction. It has the matrix form:

$P(\phi)=\begin{bmatrix}

#### $\phi$ is real number
#### Z-gate is a special case of the P-gate, with $\phi=\pi$

### The S-gates

#### This is a P-gate with $\phi=\frac{\pi}{2}$. It does a quarter-turn around the Bloch sphere. The S-gate is **not** its own inverse.

#### Again  $S^\dagger$ is P-gate with $\phi=-\frac{\pi}{2}$

#### $S=\begin{bmatrix}1 & 0 \\0 & e^{\frac{\imath\pi}{2}}\end{bmatrix}$

#### $S^\dagger=\begin{bmatrix}1 & 0 \\0 & e^{\frac{-\imath\pi}{2}}\end{bmatrix}$

#### $S^\dagger S \neq I$

#### S gate is also known as $\sqrt{Z} gate$ because $SS\left.|q\right\rangle = Z\left.|q\right\rangle$

#### Applying S-Gate

## The U-gate

#### The U-gate is the most general of all single-qubit quantum gates. It is a parametrised gate of the form:


#### $U(\theta,\phi,\lambda)=\begin{bmatrix}\cos(\frac{\theta}{2}) & -e^{\imath\lambda}\sin(\frac{\theta}{2})\\e^{\imath\phi}\sin(\frac{\theta}{2}) & e^{\imath(\phi+\lambda)}\cos(\frac{\theta}{2})\end{bmatrix}$



#### H-Gate using U-Gate

#### $U(\frac{\pi}{2},0,\pi)=\frac{1}{\sqrt{2}}(\begin{bmatrix}1 & 1 \\ 1 & -1\end{bmatrix})=H$

#### P-Gate using U-Gate


#### $U(0,0,\lambda)=\begin{bmatrix}1 & 0 \\ 0 & e^{\imath\lambda}\end{bmatrix}=P$

```python
import numpy as np
qc=QuantumCircuit(1)
# initial_state=[1,1]/np.sqrt(2)
qc.u(np.pi/3,0,np.pi/5,0)
qc.save_statevector()
qc.draw()
```
