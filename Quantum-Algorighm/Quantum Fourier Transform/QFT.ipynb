# What is QFT?
### QFT is effectively a change of basis from the computational basis to the Fourier Basis


### Eg: 1-qubit : Computational basis state are $\{|0\rangle,|1\rangle\}$
![zeroone](./101state.png)

### Fourier Basis for one Qubit is $\{ |+\rangle, |-\rangle \}$

![fourierbasis](./fourierbassis.png)

#### The quantum Fourier transform (QFT) is the quantum implementation of the discrete Fourier transform over the amplitudes of a wavefunction

#### The discrete Fourier transform acts on a vector $(x_0,\dots,x_{N-1})$ and maps it to the vector $(y_0,\dots,y_{N-1})$ according to the formula

### $y_k=\frac{1}{\sqrt{N}} \sum\limits_{j=0}^{N-1} x_jw^{j k}_N$


### where $w^{\jmath k}_N=e^{2\pi\imath\frac{jk}{N}}$



#### Similarly, the quantum Fourier transform acts on a quantum state $|X\rangle=\sum\limits_{j=0}^{N-1} x_j|j\rangle$ and maps it to the quantum state $|Y\rangle=\sum\limits_{k=0}^{N-1} x_k|k\rangle$ according to the formula

### $y_k=\frac{1}{\sqrt{N}} \sum\limits_{j=0}^{N-1} x_jw^{j k}_N$

#### **. Note that only the amplitudes of the state were affected by this transformation.**

#### This can also be expressed as the map:

### $|j\rangle\rightarrow \frac{1}{\sqrt{N}} \sum\limits_{j=0}^{N-1} w^{j k}_N |k\rangle$


### Or the unitary matrix

### $U_{QFT}=\frac{1}{\sqrt{N}} \sum\limits_{j=0}^{N-1}\sum\limits_{k=0}^{N-1} w^{j k}_N |k\rangle\langle j|$


# Intuition

#### The quantum Fourier transform (QFT) transforms between two bases, the computational (Z) basis, and the Fourier basis. **The H-gate is the single-qubit QFT**, and it transforms between the Z-basis states $|0\rangle$  and $|1\rangle$ to the X-basis states $|+\rangle$  and $|-\rangle$ . **The QFT is simply the function that transforms between these bases.**

### $|State in Computational Basis\rangle \xrightarrow{QFT}|State in Fourier Basis\rangle$

### $QFT|x\rangle= |\widetilde{x}\rangle$

### $|j\rangle\rightarrow \frac{1}{\sqrt{N}} \sum\limits_{j=0}^{N-1} w^{j k}_N |k\rangle$

#### What's it's doing is it's taking a state that represents an integer $j$ for an instance on three qubits the state $7$ would be $111$ . And we map the integer $j$ to a superposition that is uniform in magnitude but **Varies in Phase**. And we have a superposition  over all the computational basis $|k\rangle$




![counting-zbasis](./zbasis-counting.gif)

#### the leftmost qubit flips with every increment in the number, the next with every 2 increments, the third with every 4 increments, and so on. In the Fourier basis, we store numbers using different rotations around the Z-axis:


![rotation-zbasis](./fourierbasis-counting.gif)


#### What it actually means that  in computational basis we store number using the state $|0\rangle and |1\rangle$

### if we have $4 Qubit$ states and we have state $|0010\rangle$ in decimal it means $2$ means if we use **Fourier Transform** then the $2nd Qubit$ will take  $2 steps$ at bloch sphere  around the Computational Basis or Z-basis.

### The number we want to store dictates the angle at which each qubit is rotated around the Z-axis. In the state $|\widetilde{ 0}\rangle$, all qubits are in the state $|+\rangle$. As seen in the example above, to encode the state on 4 qubits, we rotated the leftmost qubit by $\frac{5}{2^n}=\frac{5}{16}$ full turns ( $1\times \frac{\pi}{2}+\frac{\pi}{4}$ radians). The next qubit is turned double this  ( $2\times \frac{\pi}{2}+\frac{\pi}{2} $ radians), for the $3rd$ qubit ($4\times \frac{\pi}{2}+\frac{\pi}{2}$ ) , for the $4th$ qubit ($8\times \frac{\pi}{2}+\pi$ ).


## How this Helps us 
### **Quantum Superposition:** One of the fundamental principles of quantum mechanics is superposition, which allows a quantum system to be in multiple states simultaneously. By encoding a particular number using Fourier transform, we can represent it as a superposition of different states. This enables parallel processing and allows us to perform computations on multiple values simultaneously, providing potential speedup in certain algorithms.

### **Quantum Fourier Transform (QFT):** The Quantum Fourier Transform is a quantum algorithm that efficiently computes the discrete Fourier transform (DFT) of a quantum state. By applying QFT to a qubit encoding a particular number, we can extract the frequency components of that number, obtaining the distribution of amplitudes across different states. This information can be useful for various applications, such as solving the period-finding problem in Shor's algorithm or analyzing the energy spectrum in quantum simulation.

## Building Quantum Circuit

### n qubits: 1 qubit : $\{ |0\rangle, |1\rangle \}$ 2 basis state

### 2 qubit : $\{|00\rangle,|01\rangle,|10\rangle,|11\rangle \}$ 4 basis state

### $N-qubits$ : $2^n$ basis states $N=2^n$

### $|\widetilde{x}\rangle=QFT|x\rangle$
### $|\widetilde{x}\rangle$ Fourier Basis
### $|x\rangle$ Computational Basis

#### Quantum Fourier Transform is exectly the **Inverse Discrete Fourier Transform**
### $|\widetilde{x}\rangle=QFT|x\rangle=\frac{1}{\sqrt{N}}\sum\limits_{y=0}^{N-1} e^{\frac{2\pi ixy}{N}}|y\rangle$


### 1 Qubit case: $[N=2^1=2]$
### $QFT|x\rangle=\frac{1}{\sqrt{2}}\sum\limits_{y=0}^{2-1} e^{\frac{2\pi ixy}{2}}|y\rangle$
### $=\frac{1}{\sqrt{2}}[e^{ \frac{2\pi i x \times 0}{2}}|0\rangle + e^{ \frac{2\pi i x \times 1}{2}}|1\rangle]$
### $=\frac{1}{\sqrt{2}}[|0\rangle+e^{\pi i x \times 1}|1\rangle]$

### $QFT|0\rangle=\frac{1}{\sqrt{2}}[|0\rangle+e^{\pi i 0 \times 1}|1\rangle]$
### $=\frac{1}{\sqrt{2}}[|0\rangle+|1\rangle]=|+\rangle$

### $QFT|1\rangle=\frac{1}{\sqrt{2}}[|0\rangle+e^{\pi i 1 \times 1}|1\rangle]$
### $=\frac{1}{\sqrt{2}}[|0\rangle-|1\rangle]=|-\rangle$


### Using Euler's Form $e^{i\pi}=\cos{\pi}+i\sin{\pi}$




## Generalize 1-Qubit QFT

#### QFT operator acts on a single qubit state

### $|\psi\rangle=\alpha|0\rangle+\beta|1\rangle$

#### In this case $x_0=\alpha,x_1=\beta N=2$

### $y_0=\frac{1}{\sqrt{2}}(\alpha exp(2\pi i \frac{0\times0}{2}))+(\beta exp(2\pi i \frac{1\times0}{2}))=\frac{1}{\sqrt{2}}(\alpha+\beta)$

### $y_1=\frac{1}{\sqrt{2}}(\alpha exp(2\pi i \frac{0\times1}{2}))+(\beta exp(2\pi i \frac{1\times 1}{2}))=\frac{1}{\sqrt{2}}(\alpha-\beta)$


### $U_{QFT}|\psi\rangle=\frac{1}{\sqrt{2}}(\alpha+\beta)|0\rangle+\frac{1}{\sqrt{2}}(\alpha-\beta)|1\rangle$


#### This operation is exectly the result of applying the **Hadamard** operator $(H)$

### $H=\frac{1}{\sqrt{2}}\begin{bmatrix}1 & 1\\ 1& -1\end{bmatrix}$

## General Circuit for QFT

### Using Inverse DFT

```python
def qft_circuit(circuit,nqubits):
    qft_qc_circuit=circuit

    for qubit in range(nqubits):
        
        qft_qc_circuit.h(qubit)
        for otherqubit in range(qubit+1,nqubits):
            qft_qc_circuit.cp(np.pi/(2**(otherqubit-qubit)),otherqubit,qubit)
            # print(qubit,otherqubit)
    return qft_qc_circuit
```

```python
qc=QuantumCircuit(3)
qft_circuit(qc,3).draw('mpl')
```

### make inverse QFT using Swap gate

```python

def qft_rotations(circuit, n):
    if n == 0:
        return circuit
    n -= 1
    circuit.h(n)
    for qubit in range(n):
        circuit.cp(np.pi/2**(n-qubit), qubit, n)

    qft_rotations(circuit, n)
```

```python
def swap_registers(circuit, n):
    for qubit in range(n//2):
        circuit.swap(qubit, n-qubit-1)
    return circuit

def qft(circuit, n):
    qft_rotations(circuit, n)
    swap_registers(circuit, n)
    return circuit
```

```python
def inverse_qft(circuit, n):
    """Does the inverse QFT on the first n qubits in circuit"""
    # First we create a QFT circuit of the correct size:
    qft_circ = qft(QuantumCircuit(n), n)
    # Then we take the inverse of this circuit
    invqft_circ = qft_circ.inverse()
    # And add it to the first n qubits in our existing circuit
    circuit.append(invqft_circ, circuit.qubits[:n])
    return circuit.decompose() # .decompose() allows us to see the individual gates
```
