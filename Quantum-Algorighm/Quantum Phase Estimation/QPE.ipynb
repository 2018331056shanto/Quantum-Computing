## A unitary matrix has Eigenvalues in the form of $e^{i\theta}$. And it has Eigenvectors that form an Orthonormal Basis

### $U|\psi\rangle=e^{i\theta_\psi}|\psi\rangle$

#### Can we extract $\theta_\psi$ given the ability to prepare $|\psi\rangle$ adn the ability to apply $U$


#### Given a unitary operator $U$ , the algorithm estimates $\theta$ in $U|\psi\rangle=e^{2\pi i\theta}|\psi\rangle$ . Here $|\psi\rangle$ is an eigenvector and $e^{2\pi i\theta}$ is the corresponding eigenvalue. Since $U$ is unitary, all of its eigenvalues have a norm of 1.

### $\frac{1}{\sqrt{2}}(|0\rangle+|1\rangle)$
### $ then P(0)=\frac{1}{2},P(1)=\frac{1}{2}$ 

#### Now if we apply a **Global Phase** $e^{i\theta}$

### $e^{i\pi/2}\frac{1}{\sqrt{2}}(|0\rangle+|1\rangle)$

### then probability of measuring both $0$ and $1$ is $\frac{1}{2}$

### $|\frac{e^{i\pi/}}{\sqrt{2}}|^2=\frac{1}{2}$

#### Now we have applied a a global phase here, but still now the probability of finding $0$ and $1$ is same as before. This is the difficulty of applying a global phase and being able to measure it out.

### If the global phase was applied by some unitary operator then we can use a **Quantum Phase Estimation** algorithm to find the global phase


## Why do we even care Global Phase?

## Overview 

#### The general quantum circuit for phase estimation is shown below. The top register contains $t$ 'counting' qubits, and the bottom contains qubits in the state $|\psi\rangle$:

![circuitqpe](./qpe_tex_qz.png)

## Intution

![fourier](./qpe_intuition.jpg)
#### In the Fourier basis the topmost qubit completes one full rotation when counting between $ 0 $  and $2^t$ . To count to a number,$x$  between  $ 0$ and $2^t$ , we rotate this qubit by $\frac{x}{2^t}$ around the z-axis. For the next qubit we rotate by $\frac{2x}{2^t}$, then $\frac{4x}{2^t}$ for the third qubit.

#### When we use a qubit to control the $U$-gate, the qubit will turn proportionally to the phase $e^{2i\pi\theta}$ . We can use successive $U$ -gates to repeat this rotation an appropriate number of times until we have encoded the phase theta as a number between $0$ and $2^t$ in the Fourier basis.

## QPE(Quantum Phase Estimation) trick 

#### I have the ability to prepare $|\psi\rangle$ and the ability to appy $U$

![QPE](./qpe.png)

### So first we have a qubit in $|0\rangle$ state and then we apply $H$(Hadamard Gate). We also have $|\psi\rangle$. Next we apply **controlled U** gate or **CP(control phase-gate)** on $|\psi\rangle$

![qpestep](./qpestep.png)

## Steps in QPE

## Step 0:

### $|0\rangle|\psi\rangle$

## Step 1:
#### Apply Hadamard Gate to $|0\rangle$

### $\frac{1}{\sqrt{2}}(|0\rangle+|1\rangle)|\psi\rangle$

### $=\frac{1}{\sqrt{2}}(|0\rangle|\psi\rangle+|1\rangle|\psi\rangle)$

## Step 2:

#### Apply Control Unitary gate in $|\psi\rangle$ and it apply a phase into $|\psi\rangle$

### $\frac{1}{\sqrt{2}}(|0\rangle|\psi\rangle+|1\rangle e^{i\theta_\psi}|\psi\rangle)$

#### Phase Kick-Back
###  $\frac{1}{\sqrt{2}}(|0\rangle+e^{i\theta_\psi}|1\rangle )|\psi\rangle$

## Step 3:
#### Apply Hadamard gate into $1st$ qubit

###  $\frac{1}{\sqrt{2}}(\frac{1}{\sqrt{2}}(|0\rangle+|1\rangle)|\psi\rangle+e^{i\theta_\psi}\frac{1}{\sqrt{2}}(|0\rangle-|1\rangle)|\psi\rangle) $

### $=\frac{1}{2}[|0\rangle(1+e^{i\theta_\psi})+|1\rangle(1-e^{i\theta_\psi})]|\psi\rangle$


#### Now probability of Measuring $0$ in first qubit:

### $|\frac{1}{2}1+e^{i\theta_\psi}|^2$

#### Now probability of Measuring $1$ in first qubit:


### $|\frac{1}{2}1-e^{i\theta_\psi}|^2$



#### Now if 
### $\theta_\psi=0\degree$ then $prob(0),prob(1)=\{ 0.9999, 7.6\times10^-5\}$

### $\theta_\psi=10\degree$ then $prob(0),prob(1)=\{ 0.9924, 0.007596\}$

#### All it's telling that  the difference between the probability of measuring zero or one, We've **encoded that phase**. In other words, we've taken that phase information and turned it into an amplitude that we can measure


### **To get the precise  $\theta_\psi$ we have to use more Qubit**
![nqubits](./morequbits.png)

## Step 0:
#### We have $n$ qubits and $|\psi\rangle$ all in $|0\rangle$ state
### $ |0\rangle^{\otimes n} |\psi\rangle$

## Step 1:
#### We apply **Hadamard Gate** to all $n$ qubits

### $(\frac{1}{\sqrt{2}})^n (|0\rangle+|1\rangle)^{\otimes n}|\psi\rangle$

### $U^{2^x}|\psi\rangle=U^{2^x-1} U^x|\psi\rangle$
### $=U^{2^x-1} e^{i\theta_\psi}|\psi\rangle$
### $=e^{i\theta_\psi}U^{2^x-1}|\psi\rangle$
### $=e^{i\theta_\psi}U^{2^x-2} U^x|\psi\rangle$
### $=e^{i\theta_\psi}U^{2^x-2} e^{i\theta_\psi}|\psi\rangle$
### $=e^{i\theta_\psi}e^{i\theta_\psi}U^{2^x-2} |\psi\rangle$

#### If we continue to do this process 


## Step Final:
#### Applying all the $n$ controlled operations $CU^{2x}$ with $0\le x \le n-1$ and using the relations
### $|0\rangle \otimes |\psi\rangle +|1\rangle \otimes e^{2\pi i \theta}|\psi\rangle=(|0\rangle+e^{2\pi i\theta }|1\rangle)\otimes |\psi\rangle$
#### Now
### $(\frac{1}{\sqrt{2}})(|0\rangle+e^{i\theta_\psi 2^{n-1}}|1\rangle)\otimes (|0\rangle+e^{i\theta_\psi 2^{n-2}}|1\rangle)\otimes \dots \otimes  (|0\rangle+e^{i\theta_\psi 2^{1}}|1\rangle) \otimes (|0\rangle+e^{i\theta_\psi 2^{0}}|1\rangle)|\psi\rangle$

### $=\frac{1}{\sqrt{2^n}}\sum\limits_{k=0}^{2^n-1} e^{2\pi i k\theta}|k\rangle \otimes |\psi\rangle$


#### where $k$ denotes the integer representation of n-bit binary numbers.



#### Now if we look at $QFT$ 
### $|\widetilde{x}\rangle =\frac{1}{\sqrt{N}}(|0\rangle+e^{\frac{2\pi ix}{2^1}}|1\rangle)\otimes (|0\rangle+e^{\frac{2\pi ix}{2^2}}|1\rangle)\otimes (|0\rangle+e^{\frac{2\pi ix}{2^3}}|1\rangle)+\otimes \dots +\otimes (|0\rangle+e^{\frac{2\pi ix}{2^n}}|1\rangle)$


### **QPE same as QFT except $\theta_\psi \rightarrow \frac{\theta_\psi 2\pi}{2^n}$**


## Inverse Fourier Transform 

### $\frac{1}{\sqrt{2^n}}\sum\limits_{k=0}^{2^n-1} e^{2\pi i k\theta}|k\rangle \otimes |\psi\rangle \xrightarrow{QFT_n^-1} \frac{1}{2^n}\sum\limits_{x=0}^{2^n-1}\sum\limits_{k=0}^{2^n-1}e^{-\frac{2\pi i k}{2^n}(x-2^n\theta)}|x\rangle \otimes |\psi\rangle$


## Measurement 

#### The above expression peaks near $x=2^n\theta$. For the case when $2^n\theta$ is an integer, measuring in the computational basis gives the phase in the auxiliary register with high probability: 
### $=|2^n\theta\rangle \otimes |\psi\rangle$

![inverse-qft](./qft-1.png)

### So we have to perform a **Inverse QFT** to undo the QFT that happend and expect to see the state $|2^{n}\theta_\psi\rangle$

### The reason we took multiple qubit  becasue in $1$ qubit we didn't have enough precision to do our measurement but in this case we've multiplied that phase by $2^n$. So, instead of needing to do lots of shots what we've done is insted multiply the phase that we are looking for by $2^n$ which gives us more precision

## Example : T-gate

####  $T$ -gate adds a phase of $e^{\frac{i\pi}{4}}$ to the state $|1\rangle$ :

### $T|1\rangle=\begin{bmatrix}1 & 0\\0 &e^{\frac{i\pi}{4}} \end{bmatrix}\begin{bmatrix}0 \\1 \end{bmatrix}=e^{\frac{i\pi}{4}}|1\rangle$

#### $QPE$ will give us $\theta$ where

### $T|1\rangle=e^{2i\pi\theta}|1\rangle$

## 3. Example: Getting More Precision


## Phase $\theta=\frac{1}{3}$

```python
# Create and set up circuit
qpe2 = QuantumCircuit(4, 3)

# Apply H-Gates to counting qubits:
for qubit in range(3):
    qpe2.h(qubit)

# Prepare our eigenstate |psi>:
qpe2.x(3)

# Do the controlled-U operations:
angle = 2*np.pi/3
repetitions = 1
for counting_qubit in range(3):
    for i in range(repetitions):
        qpe2.cp(angle, counting_qubit, 3);
    repetitions *= 2

# Do the inverse QFT:
qpe2 = qpe2.compose(QFT(3, inverse=True), [0,1,2])

# Measure of course!
for n in range(3):
    qpe2.measure(n,n)

qpe2.draw('mpl')
```

```python
# Let's see the results!
aer_sim = Aer.get_backend('aer_simulator')
shots = 4096
t_qpe2 = transpile(qpe2, aer_sim)
results = aer_sim.run(t_qpe2, shots=shots).result()
answer = results.get_counts()

plot_histogram(answer)
```

#### We are expecting the result $\theta=0.3333...$ , and we see our most likely results are 010(bin) = 2(dec) and 011(bin) = 3(dec). These two results would tell us that  $\theta=0.25$(off by 25%) and  $\theta=0.375$ (off by 13%) respectively. The true value of $\theta$ lies between the values we can get from our counting bits, and this gives us uncertainty and

### To get more precision we simply add more counting qubits. We are going to add two more counting qubits

```python
# Create and set up circuit
qpe3 = QuantumCircuit(6, 5)

# Apply H-Gates to counting qubits:
for qubit in range(5):
    qpe3.h(qubit)

# Prepare our eigenstate |psi>:
qpe3.x(5)

# Do the controlled-U operations:
angle = 2*np.pi/3
repetitions = 1
for counting_qubit in range(5):
    for i in range(repetitions):
        qpe3.cp(angle, counting_qubit, 5);
    repetitions *= 2

# Do the inverse QFT:
qpe3 = qpe3.compose(QFT(5, inverse=True), range(5))

# Measure of course!
qpe3.barrier()
for n in range(5):
    qpe3.measure(n,n)

qpe3.draw('mpl')
```

```python
# Let's see the results!
aer_sim = Aer.get_backend('aer_simulator')
shots = 4096
t_qpe3 = transpile(qpe3, aer_sim)
results = aer_sim.run(t_qpe3, shots=shots).result()
answer = results.get_counts()

plot_histogram(answer)
```

### The two most likely measurements are now 01011 (decimal 11) and 01010 (decimal 10). Measuring these results would tell us  is:

### $\theta=\frac{11}{2^4}=.344 or \theta=\frac{10}{2^5}=.313$

#### These two results differ from $\frac{1}{3}$ by 3% and 6% respectively. A much better precision!
